##  visualisation of number of visitors (adults,children and babies) visited by each country

```python
cancel_by_adults = df[(df['is_canceled']==0) & (df['adults'])]['hotel'].value_counts().reset_index()
cancel_by_adults.columns = ['hotel', 'no of guest']
# cancel_by_adults

cancel_by_children = df[(df['is_canceled']==0) & (df['children'])]['hotel'].value_counts().reset_index()
cancel_by_children.columns = ['hotel', 'no of guest']
# cancel_by_children

cancel_by_babies = df[(df['is_canceled']==0) & (df['babies'])]['hotel'].value_counts().reset_index()
cancel_by_babies.columns = ['hotel', 'no of guest']
# cancel_by_adults

# visualize them all

plt.figure(figsize=(14,6))
sns.lineplot(data=cancel_by_adults, x=cancel_by_adults["hotel"],y=cancel_by_adults["no of guest"])
sns.lineplot(data=cancel_by_children, x=cancel_by_children["hotel"],y=cancel_by_children["no of guest"])
sns.lineplot(data=cancel_by_babies, x=cancel_by_babies["hotel"],y=cancel_by_babies["no of guest"])

plt.show()
```
